# Maximum Likelihood Estimation (MLE) Plotting in Python

## Objective
The goal of this project was to plot the likelihood, log-likelihood, and negative log-likelihood functions for a coin toss experiment. We also wanted to calculate and show the Maximum Likelihood Estimate (MLE) of the probability of heads, \( \theta \), on the plots.

## Dataset Used
- Number of Heads (\( n_{heads} \)): 3
- Number of Tails (\( n_{tails} \)): 2
- Total Tosses (\( n \)): 5

## Key Concepts
- **Likelihood Function**: \( L(\theta) = \theta^{n_{heads}} (1-\theta)^{n_{tails}} \)
- **Log-Likelihood Function**: \( \log L(\theta) = n_{heads} \log \theta + n_{tails} \log (1-\theta) \)
- **Negative Log-Likelihood Function**: \( -\log L(\theta) \)
- **MLE**: The value of \( \theta \) maximizing the likelihood, calculated as:
  \[
  \hat{\theta} = \frac{n_{heads}}{n_{heads} + n_{tails}}
  \]

## Implementation Details
- Used `numpy` to generate theta values and calculate likelihoods.
- Used `matplotlib` to plot the functions.
- Marked the MLE as a red dashed vertical line on all the plots.
- Plotted three separate graphs: Likelihood, Log-Likelihood, and Negative Log-Likelihood.
- Made sure each graph starts from (0, 0) on axes where applicable.

## Example Code Snippet



```python
import numpy as np
import matplotlib.pyplot as plt
n_heads = 3
n_tails = 2
n = n_heads + n_tails
theta = np.linspace(0.001, 0.999, 500)
```

```python
likelihood = (theta**n_heads) * ((1 - theta)**n_tails)
log_likelihood = n_heads * np.log(theta) + n_tails * np.log(1 - theta)
neg_log_likelihood = -log_likelihood
mle = n_heads / (n_heads + n_tails)

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.plot(theta, likelihood, color="blue")
plt.axvline(mle, color="red", linestyle="--", label=f"MLE ={mle:.2f}")
plt.title("Likelihood Function L()")
plt.xlabel("")
plt.ylabel("L()")
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(theta, log_likelihood, color="green")
plt.axvline(mle, color="red", linestyle="--", label=f"MLE ={mle:.2f}")
plt.title("Log-Likelihood Function log L()")
plt.xlabel("")
plt.ylabel("log L()")
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(theta, neg_log_likelihood, color="purple")
plt.axvline(mle, color="red", linestyle="--", label=f"MLE ={mle:.2f}")
plt.title("- Log-Likelihood Function")
plt.xlabel("")
plt.ylabel("- log L()")
plt.legend()

plt.tight_layout()
plt.show()
```
